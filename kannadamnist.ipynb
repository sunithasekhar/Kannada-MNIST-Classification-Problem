```python
#kannada_digit MNIST prob;em
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix,roc_auc_score, auc
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn import tree
from sklearn import metrics
```

```python
#load MNIST dataset

data_X_train=np.load('/content/X_kannada_MNIST_train.npz')
X_train=data_X_train['arr_0']
data_X_test=np.load('/content/X_kannada_MNIST_test.npz')
X_test=data_X_test['arr_0']

data_y_train=np.load('/content/y_kannada_MNIST_train.npz')
y_train=data_y_train['arr_0']
data_y_test=np.load('/content/y_kannada_MNIST_test.npz')
y_test=data_y_test['arr_0']
```

```python
X_train=data_X_train['arr_0']
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# reshape the data
X_train_reshape=X_train.reshape(60000,28*28)
X_test_reshape=X_test.reshape(10000,28*28)
print(X_train_reshape.shape)
print(X_test_reshape.shape)

```

```python
# Standaridise the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled=scaler.fit_transform(X_train_reshape)
X_test_scaled=scaler.transform(X_test_reshape)
print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

```python
#apply PCA to 10 components
from sklearn.decomposition import PCA
pca=PCA(n_components=10)

X_train_pca=pca.fit_transform(X_train_scaled)
print(X_train_pca.shape)

```

```python
#appy pca to X_test_scaled
X_test_pca=pca.fit_transform(X_test_scaled)
print(X_test_pca.shape)
```

```python
# model1:apply the decisiontreemodel
dectree=DecisionTreeClassifier()
dectree.fit(X_train_pca,y_train)
```

```python
y_predict_dectree=dectree.predict(X_test_pca)
```

```python
print(metrics.classification_report(y_test,y_predict_dectree))
print(metrics.confusion_matrix(y_test,y_predict_dectree))
```

```python
print("Decision Tree Accuracy:", metrics.accuracy_score(y_test, y_predict_dectree))
```

```python
# Apply the model for different values of pca (pca=10,15,20,25,30)
#model1:Decisiontreeclassifier
pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  dectree=DecisionTreeClassifier()
  dectree.fit(X_train_pca,y_train)
  y_predict_dectree=dectree.predict(X_test_pca)

  print(metrics.classification_report(y_test,y_predict_dectree))
  print(metrics.confusion_matrix(y_test,y_predict_dectree))
  print("Decision tree accuracy", metrics.accuracy_score(y_test,y_predict_dectree))
  print("-" *75)

```

```python
# prompt: write above codes for random forest

#model2:Apply RandomForestClassifier
pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42) # You can adjust n_estimators
  rf_classifier.fit(X_train_pca, y_train)
  y_pred_rf = rf_classifier.predict(X_test_pca)

  print(metrics.classification_report(y_test, y_pred_rf))
  print(metrics.confusion_matrix(y_test, y_pred_rf))
  print("Random Forest Accuracy:", metrics.accuracy_score(y_test, y_pred_rf))
  print("-" * 75)
```

```python
#Model:2 Random Forest
# Apply the model for different values of pca (pca=10,15,20,25,30)

pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42)
  rf_classifier.fit(X_train_pca,y_train)
  y_predict_rf=rf_classifier.predict(X_test_pca)

  print(metrics.classification_report(y_test,y_predict_rf))
  print(metrics.confusion_matrix(y_test,y_predict_rf))
  print("Decision tree accuracy", metrics.accuracy_score(y_test,y_predict_rf))
  print("-" *75)
```

```python
# mode3:3 Naive Bayes model
# Apply the model for different values of pca (pca=10,15,20,25,30)

pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  gnb=GaussianNB()
  gnb.fit(X_train_pca,y_train)
  y_predict_gnb=gnb.predict(X_test_pca)


  print(metrics.classification_report(y_test,y_predict_gnb))
  print(metrics.confusion_matrix(y_test,y_predict_gnb))
  print("Gaussian Naiva Bayes accuracy", metrics.accuracy_score(y_test,y_predict_gnb))
  print("-" *75)
```

```python
#model:4 KNN Classifier

# Apply the model for different values of pca (pca=10,15,20,25,30)

pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  knn_classifier=KNeighborsClassifier(n_neighbors=5)
  knn_classifier.fit(X_train_pca,y_train)
  y_predict_knn=knn_classifier.predict(X_test_pca)

  print(metrics.classification_report(y_test,y_predict_knn))
  print(metrics.confusion_matrix(y_test,y_predict_knn))
  print(" KNN accuracy", metrics.accuracy_score(y_test,y_predict_knn))
  print("-" *75)
```

```python
#model:5 SVM
# Apply the model for different values of pca (pca=10,15,20,25,30)

pca_values=[10,15,20,25,30]

for n_components in pca_values:
  print(f"Applying PCA with {n_components} components")
  pca=PCA(n_components=n_components)
  X_train_pca=pca.fit_transform(X_train_scaled)
  X_test_pca=pca.transform(X_test_scaled)

  svm_classifier=SVC(kernel='linear',C=1)
  svm_classifier.fit(X_train_pca,y_train)
  y_predict_svm=svm_classifier.predict(X_test_pca)

  print(metrics.classification_report(y_test,y_predict_svm))
  print(metrics.confusion_matrix(y_test,y_predict_svm))
  print("SVM accuracy", metrics.accuracy_score(y_test,y_predict_svm))
  print("-" *75)
```
